```python
import pylab
import nltk
%pylab inline
def zipf(text):
    fd = nltk.FreqDist(text)
    rank = 1
    freqs = []
    ranks = []
    for sample, count in fd.most_common(200):
        if any(c.isalpha() for c in sample):
            freqs.append(fd.freq(sample))
            ranks.append(rank)
            rank = rank + 1
    pylab.plot(ranks, freqs)
```

```python
emma = nltk.Text(nltk.corpus.gutenberg.words('austen-emma.txt'))
zipf(emma)
```
